```python
def medianFinder(A):
    n = len(A)
    B =[]
    
    for i in range((n//2)):
        maxi = max(A)
        B.append(maxi)
        A.remove(maxi)
    
    if n%2==0:
        return (max(A)+min(B))//2
    else:
        return max(A)
    
        
            


A = []
for i in range(100001):
    A.append(i)


startT = time.time()
print(medianFinder(A))
endT = time.time()
print("Time required: ", endT-startT)

B=[]
for i in range(100001):
    B.append(i)

startT = time.time()
print(np.median(B))
endT= time.time()
print("Time required: ",endT -startT)

#Sorting it using the inbuilt sort function to find the median. Mostly the inbuilt function uses merge sort which takes nlogn time to complete, and then constant time to retrive the element from the array.
```

As we can clearly see that the median finding algorithm we implemented is of order ~O(n^2) and takes a lot of computational time for finding medians of larger arrays. But in the case of np.median() it is able to find the median pretty faster due to the implementation of various other algorithms such as QuickSelect and other few that make it quicker.

Question 12:

Broadcasting in NumPy

Let us understand it using an example. Let us consider two arrays A = [2,4,35,6] and B = [9,4,22,5]. Can we add it?


```python
A = np.array([2,4,35,6])
B = np.array([9,4,22,5])

print(A+B)
```

We can clearly see that we are able to add the two list (treat as matrices of 1×4). But what happens when we add two matrices of different dimensions. For example 1×4 and 1×5

```python
C = np.array([1,3,4,5])
D = np.array([7,44,6,3,5])

print(C+D)
```

We get the broadcasting error. This error is basically unequal dimensions and unability performing operations on them. Broadcasting has two rules basically.
1. To perform an operation on matrices, they should have same dimensions.
2. If not then, atleast of them should be same (not 1) and other can be 1 or the same number.
Notation (Rows×Column)
for example we can do 2×2 and 3×2 because we have '2' as same and 3 has one with it 


```python
A = np.array([1,2])
B = np.array([[1,2],[2,3],[3,4]])
print(A.shape)
print(B.shape)
print(A+B)
```

```python
def minOfArray(arr):
    sizeOfArr = len(arr)
    index = -1
    minOfArr = float('inf')
    for i in range(sizeOfArr):
        if minOfArr>arr[i]:
            minOfArr = arr[i]
            index = i

    return index

#Example to check the working of the code.
arr= np.array([1,34,4,5,53,2,1,3])
print("Index of minimum value recieved by function:",minOfArray(arr))
print("Index of minimum value recieved by inbuilt function :",np.argmin(arr))
```
